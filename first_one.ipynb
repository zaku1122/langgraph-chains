```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name: str
    message: str



def compliment_node(state: AgentState) -> AgentState:
    """ Simple node that adds a greeding message to the state."""
    name = state["name"].capitalize()

    state["message"] = state.get("message", "") + f" Amazing Job, learning L.G. {name}!"

    return state
```

```python
#Build the graph

graph = StateGraph(AgentState)

graph.add_node(
    "compliment", compliment_node
)
graph.set_entry_point("compliment")
graph.set_finish_point("compliment")

app = graph.compile()
```

```python
from IPython.display import display, Image
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"name": "Zakwanzahid", "message": ""})
```
